```python
import pandas as np
import matplotlib.pyplot as plt
```

```python
#CRIM	SuÃ§ oranÄ± (her kiÅŸi baÅŸÄ±na dÃ¼ÅŸen suÃ§ oranÄ±)
#ZN	25.000 mÂ²'den bÃ¼yÃ¼k konut arazisi oranÄ± (%)
#INDUS	Ticari olmayan arazi oranÄ± (%)
#CHAS	Charles Nehri'ne yakÄ±nlÄ±k (1 = yakÄ±n, 0 = deÄŸil)
#NOX	Azot oksit konsantrasyonu (ppm cinsinden hava kirliliÄŸi)
#RM	Ortalama oda sayÄ±sÄ± (bir konut birimi baÅŸÄ±na)
#AGE	1940â€™tan Ã¶nce inÅŸa edilen evlerin yÃ¼zdesi (%)
#DIS	Boston iÅŸ merkezlerine olan aÄŸÄ±rlÄ±klÄ± mesafe (katsayÄ± ile Ã¶lÃ§eklendirilmiÅŸ)
#RAD	Otoyollara eriÅŸilebilirlik endeksi
#TAX	Emlak vergisi oranÄ± (her 10.000$â€™lÄ±k deÄŸer baÅŸÄ±na)
#PTRATIO	Ã–ÄŸretmen baÅŸÄ±na dÃ¼ÅŸen Ã¶ÄŸrenci oranÄ±
#B	Siyahi nÃ¼fus oranÄ± (1000(Bk - 0.63)Â²; Bk: kasabadaki siyahi nÃ¼fus oranÄ±)
#LSTAT	DÃ¼ÅŸÃ¼k gelirli nÃ¼fus oranÄ± (%)
#MEDV	Medyan konut fiyatÄ± (bin $ cinsinden)
```

```python
#dt["CRIM"].fillna(dt["CRIM"].median(), inplace=True) #medyan deger ile doldurduk
#dt["ZN"].fillna(dt["ZN"].mode()[0], inplace=True) #en sÄ±k tekrar eden deger ile doldurduk
#dt["CHAS"].fillna(0, inplace=True)  # 0 ile doldurduk
#dt["INDUS"].fillna(dt["INDUS"].median(), inplace=True) #medyan deger ile doldurduk
#dt["AGE"].fillna(dt["AGE"].median(), inplace=True) #medyan deger ile doldurduk
dt["CRIM"] = dt["CRIM"].fillna(dt["CRIM"].median())  
dt["ZN"] = dt["ZN"].fillna(dt["ZN"].mode()[0])  
dt["CHAS"] = dt["CHAS"].fillna(0)  
dt["INDUS"] = dt["INDUS"].fillna(dt["INDUS"].median())  
dt["AGE"] = dt["AGE"].fillna(dt["AGE"].median())  
#dt.isnull().sum()
```

```python
dt.to_csv("HousingDataNew.csv", index=False)
```
