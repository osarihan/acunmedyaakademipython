```python
import pandas as np
import matplotlib.pyplot as plt
```

```python
dt = np.read_csv("HousingData.csv")
dt
```

```python
dt.isnull().sum()
```

```python
#CRIM	Suç oranı (her kişi başına düşen suç oranı)
#ZN	25.000 m²'den büyük konut arazisi oranı (%)
#INDUS	Ticari olmayan arazi oranı (%)
#CHAS	Charles Nehri'ne yakınlık (1 = yakın, 0 = değil)
#NOX	Azot oksit konsantrasyonu (ppm cinsinden hava kirliliği)
#RM	Ortalama oda sayısı (bir konut birimi başına)
#AGE	1940’tan önce inşa edilen evlerin yüzdesi (%)
#DIS	Boston iş merkezlerine olan ağırlıklı mesafe (katsayı ile ölçeklendirilmiş)
#RAD	Otoyollara erişilebilirlik endeksi
#TAX	Emlak vergisi oranı (her 10.000$’lık değer başına)
#PTRATIO	Öğretmen başına düşen öğrenci oranı
#B	Siyahi nüfus oranı (1000(Bk - 0.63)²; Bk: kasabadaki siyahi nüfus oranı)
#LSTAT	Düşük gelirli nüfus oranı (%)
#MEDV	Medyan konut fiyatı (bin $ cinsinden)
```

```python
#dt["CRIM"].fillna(dt["CRIM"].median(), inplace=True) #medyan deger ile doldurduk
#dt["ZN"].fillna(dt["ZN"].mode()[0], inplace=True) #en sık tekrar eden deger ile doldurduk
#dt["CHAS"].fillna(0, inplace=True)  # 0 ile doldurduk
#dt["INDUS"].fillna(dt["INDUS"].median(), inplace=True) #medyan deger ile doldurduk
#dt["AGE"].fillna(dt["AGE"].median(), inplace=True) #medyan deger ile doldurduk
dt["CRIM"] = dt["CRIM"].fillna(dt["CRIM"].median())  
dt["ZN"] = dt["ZN"].fillna(dt["ZN"].mode()[0])  
dt["CHAS"] = dt["CHAS"].fillna(0)  
dt["INDUS"] = dt["INDUS"].fillna(dt["INDUS"].median())  
dt["AGE"] = dt["AGE"].fillna(dt["AGE"].median())  
#dt.isnull().sum()
```

```python
dt.to_csv("HousingDataNew.csv", index=False)
```

```python
x = dt["AGE"]
y = dt["MEDV"]
#plt.xlabel("Yaş")
#plt.ylabel("Fiyat")
plt.scatter(x, y, color='b', marker='o', alpha=0.7, label="Yaş ile Fiyat")
```

```python
plt.plot(x, y, color='g', marker='o', linestyle='-', linewidth=2, markersize=8, label="Yaş ile Fiyat line")
```
