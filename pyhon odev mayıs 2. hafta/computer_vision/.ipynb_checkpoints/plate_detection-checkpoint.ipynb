```python
from ultralytics import YOLO
```

```python
model = YOLO("yolov8n.pt")
```

```python
model.train(data="data.yaml", epochs=1, imgsz=640, batch=8) 
# epochs => tüm veri ile X adet eğitim yapılsın.
# epoch_başına_veri = toplam_veri / batch
```

## Prediction

```python
img_predict = cv2.imread("arac.jpg")
#gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
model2 = YOLO("runs/detect/train/weights/best.pt")
```

```python
results = model.predict(source=img_predict, conf=0.5) #Conf-Confident 0.5  (50%)'nin altında eminsen bunu tahmin olarak sayma.
```

```python
boxes = results[0].boxes
annotated_img = img_predict.copy()
```

```python
import matplotlib.pyplot as plt
```

```python
for box in boxes:
    x1, y1, x2, y2 = map(int, box.xyxy[0])

    h, w, _ = annotated_img.shape

    cv2.rectangle(annotated_img, (x1,y1), (x2,y2), color=(255,0,0), thickness=2)
```

```python
%matplotlib inline
plt.imshow(annotated_img)
plt.show()
```

```python
import cv2
from ultralytics import YOLO
import pytesseract
```

```python
box = boxes.xyxy[0].tolist() # Koordinatlar: [x1, y1, x2, y2]
x1, y1, x2, y2 = [int(i) for i in box]
```

```python
plate_img_cropped = annotated_img[y1:y2, x1:x2]
```

```python
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

extracted_text = pytesseract.image_to_string(plate_img_cropped, config='--psm 8')

print("Tespit Edilen Plaka Numarası:", extracted_text)
```

```python
#kesilen plakayı gösterme
cv2.imshow("Cropped Plate", plate_img_cropped)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
