```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("spam_ham_dataset.csv")
df.head(10)
```

```python
df = df[['text', 'label_num']]
df.head(3)
```

```python
# stopword kaldirma islemi. -> the, at, on, is gibi cumlede anlami degistirmeyecek kelimeler
# stem islemi (kelimeleri kokune indirme) -> runnning -> run 

import string
import nltk
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
ps = PorterStemmer()

nltk.download("stopwords")

def clean_text(text):
    text = text.lower() # tum metin kucuk harfe cevrildi

    text = text.translate(str.maketrans('','', string.punctuation)) # tum noktalama isaretleri kalkti
    word = text.split()

    words = [ word for word in word if not word.isdigit()]

    words = [ ps.stem(word) for word in words if word not in stopwords.words('english')]

    return " ".join(words)
```

```python
df['clean_text'] = df['text'].apply(clean_text)
```

```python
df.head(5)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
x = df["clean_text"]
y = df["label_num"]

tfidf = TfidfVectorizer()
X_tfidf = tfidf.fit_transform(df["clean_text"])

x_train, x_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.3, random_state=42)
```

```python
#Naive Bayes algoritmasi
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(x_train, y_train)
```
