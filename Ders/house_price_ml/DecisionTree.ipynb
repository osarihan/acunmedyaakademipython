```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("HousingData_Clean.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
x = df.drop(columns=["MEDV"])
y = df["MEDV"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor(max_depth=4, random_state=42)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
#r2_score = r2_score, yani R-Kare (R²) skoru, bir regresyon modelinin tahminlerinin doğruluğunu ölçen bir metriktir.
#Sklearn'de r2_score fonksiyonuyla hesaplanır ve genellikle regresyon problemlerinde kullanılır.
```

```python
mae = mean_absolute_error(y_test, y_pred)
mae
```

```python
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(20,10))
plot_tree(model, feature_names=x.columns, filled=True, rounded = True)
plt.title("Decision Tree")
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model2 = RandomForestRegressor(n_estimators=100, random_state=42)
model2.fit(x_train, y_train)
```

```python
y_pred2 = model2.predict(x_test)
```

```python
mae2 = mean_absolute_error(y_test, y_pred2)
mae2
```

```python
# Overfitting => Veriyi ezberlemek, öğrenmek değil.
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    "n_estimators": list(range(5,30))
}
```

```python
grid = GridSearchCV(RandomForestRegressor(), param_grid, scoring="r2")
grid.fit(x_train, y_train)
```

```python
grid.best_params_, grid.best_score_
```
