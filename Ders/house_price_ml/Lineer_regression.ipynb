```python
import pandas as pd
```

```python
df = pd.read_csv("HousingData.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df = df.fillna(df.mean()) ## her alani o alanin ortalamasiyla doldur
```

```python
df.isnull().sum()
```

```python
df.head()
df.to_csv("HousingData_Clean.csv")
```

```python
x = df.drop(columns=["MEDV"])
y = df["MEDV"]
```

```python
x
```

```python
y
```

```python
#veriyi train ve test etmek
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3,random_state=42)
```

```python
#random seed
```

```python
#lineer regresyon
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
#git egitildigin gibi su x_test verileri ile tahminler yap
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
#MAE
mae = mean_absolute_error(y_test, y_pred)
mae
#MSE
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(y_test, y_pred, label="tahminler", alpha = 0.6)
plt.x_label = "gercek fiyatlar"
plt.y_label = "tahmin fiyatlar"
plt.legend()
min_val = min(y_test.min(), y_pred.min())
max_val = max(y_test.max(), y_pred.max())
plt.plot([min_val, max_val], [min_val, max_val], color = "red", linestyle = "--")
plt.show
```
